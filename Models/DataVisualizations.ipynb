**GRAPHS FOR RMSE & R2**

```python
df = pd.DataFrame({
    'Model vs RMSE value': ['Spacy + HighGraBoost','CodeBERT + LSTM+RidgeCV','RoBERTa + CatBoost'],
        'Train': [0.59, 1.79, 0.56],

    'Test': [1.90, 1.87, 1.85]

})

# plotting graph
df.plot(x="Model vs RMSE value", y=["Train", "Test"], kind="bar")
plt.xlabel('Model')
plt.ylabel('RMSE score')
```

```python
df = pd.DataFrame({
    'Model vs R2 value': ['Spacy + HighGraBoost','CodeBERT + LSTM+RidgeCV','RoBERTa + CatBoost'],
        'Train': [0.92, 0.336, 0.93],

    'Test': [0.32, 0.33, 0.35]

})

# plotting graph
df.plot(x="Model vs R2 value", y=["Train", "Test"], kind="bar")
plt.xlabel('Model')
plt.ylabel('R2 score')
```

```python
df = pd.DataFrame({
    'R2 vs RMSE value': ['Train', 'Test'],
        'RMSE': [1.79, 1.87],

    'R2': [0.336, 0.33]

})

# plotting graph
df.plot(x="R2 vs RMSE value", y=["RMSE", "R2"], kind="bar")
plt.xlabel('Model')
plt.ylabel('Score')
plt.title('RMSE & R2 for CodeBERT+LSTM+RidgeCV')
```
